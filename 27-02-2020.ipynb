### Matplot Library

```python
# this is for line plot
from matplotlib import pyplot as plt
x = [2,4,6,8,10,12,14,16]
y = [1,2,3,5,3,7,2,9]
plt.plot(x,y)
plt.show()
```

```python
# this for bargraph
from matplotlib import pyplot as plt
x = [2,4,6,8,10,12,14,16]
y = [1,2,3,5,3,7,2,9]
plt.bar(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2,4,6,8,10,12,14,16]
y = [1,2,3,5,3,7,2,9]
plt.pie(x)
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2,4,6,8,10,12,14,16]
y = [1,2,3,5,3,7,2,9]
plt.scatter(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2,4,6,8,10,12,14,16]
y = [1,2,3,5,3,7,2,9]
plt.hist(x)
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2,4,6,8,10,12,14,16]
y = [1,2,3,5,3,7,2,9]
plt.plot(x,y)
plt.xlabel('speed')
plt.ylabel('distance')
plt.title("graph")
plt.show()
```
